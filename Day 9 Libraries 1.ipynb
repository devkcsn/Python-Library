```python
year = np.array([])
l = []
for i in range(1949,2016):
    if i%5 == 0:
        l.append(i)
year = np.array(l)
pops = [1650, 2525, 2758 , 3018,3322,3682,4061,4268,4853,5310,5735,6127,6520,7349]
plt.plot(year,pops , color = 'skyblue');
plt.title('Population over the year')
plt.xlabel('Years')
plt.ylabel('Population')

plt.xlim(1920,2000)
```

```python
year = np.array([])
l = []
for i in range(1951,2016):
    if i%5 == 0:
        l.append(i)
year = np.array(l)

pops = [2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]

death = [1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]

plt.plot(year,pops , '--' , color = (255/255, 100/255 , 255/255));
plt.plot(year,death ,'-' ,color = (1,.8, 84/255));

plt.xlabel('Years');
```

```python
fig, axes = plt.subplots(1,3, figsize = (16,5))

ax1,ax2,ax3 = axes

x=np.linspace(1,10,500)
y1 = x
y2 = np.exp(x)
y3 = np.log(x)

ax1.plot(x,y1,'--', c ='r' ,linewidth = 4 );
ax2.plot(x,y2, c ='g' ,linewidth = 4 );
ax3.plot(x,y3, c ='b' ,linewidth = 4 );
```

```python
fig, axes = plt.subplots(1,2, figsize = (16,5))
ax1,ax2 = axes

year = np.array([])
l = []
for i in range(1951,2016):
    if i%5 == 0:
        l.append(i)
year = np.array(l)

pops = [2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]

death = [1.2,1.1,1.2,2.1,2.0,2.3,1.8,1.9,2.6,1.6,2.4,2.4,4.0]


ax1.plot(year,pops , '--' , color = (255/255, 100/255 , 255/255));
ax2.plot(year,death ,'--' ,color = (1,.8, 84/255));
plt.xlabel('Years');
plt.ylabel('Death');
```

```python
films = ['Tanhaji' , 'Baghi 3' , 'Love Aaj Kal' , 'Street Dancer 3']
box_office = [367.65, 137.05,10,97]


plt.bar(films,box_office, color = 'cyan');


plt.grid(color = 'Black' , alpha = 0.21 , axis = 'y' , linewidth = 1.0,linestyle = ':');
plt.title('Film Collections' , color = "b");
plt.xlabel('Movies');
plt.ylabel('Collections (crore , $)');
```

```python
x1 , y1 = [1,3,5,7,9], [5,2,7,8,2]
x2 , y2 = [2,4,6,8,10], [8,6,2,5,6]
plt.bar(x1,y1, color = 'cyan' , label = 'Example one');
plt.bar(x2,y2, color = 'orange' , label = 'Example two');

plt.legend(loc = 'lower left')
plt.xlabel('Bar number');
plt.ylabel('Bar Height');

plt.title('Randomly Generated');
```

```python
fig, axes = plt.subplots(1,2, figsize = (16,5))
ax1,ax2 = axes

x1 , y1 = [1,3,5,7,9], [5,2,7,8,2]
x2 , y2 = [2,4,6,8,10], [8,6,2,5,6]
ax1.bar(x1,y1, color = 'cyan' );

plt.xlabel('bar number');
plt.ylabel('Bar Height');

ax2.bar(x2,y2, color = 'blue');

plt.xlabel('Bar number');
plt.ylabel('Bar Height');

plt.title('Randomly Generated');
```

```python
#Stacked bar chart

years = ['2016', '2017' , '2018','2019']
rice = [57,50,47,30]
wheat = [43,50,53,70]
mustard = [10,16,13,15]

plt.xlabel("Years")
plt.bar(years , rice , label = "Rice")
plt.bar(years ,wheat , bottom=rice, label = 'Wheat' ,color ="cyan");
plt.bar(years ,mustard,bottom=wheat and rice,label= 'Mustard' , color = "skyblue");
plt.title(" Hello See my chart");
```
