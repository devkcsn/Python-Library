```python
data = [33,52,12,17,42,48]
l = ['C' , 'C++' , 'Java' , 'Mojo' , 'Python' , 'Javascript']
exp_list = [0,0.08,0,0,0.099,0]
plt.pie(data , labels = l , autopct = '%.2f %%' , explode = exp_list, counterclock = False , wedgeprops= {'width':0.58} ,shadow = True , radius = 2);


data = [33,52,12,17,42,48]
l = ['C' , 'C++' , 'Java' , 'Mojo' , 'Python' , 'Javascript']
plt.pie(data , labels = l , autopct = '%.2f %%', counterclock = False , startangle = 0,radius = 0.9);
```

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.stackplot(days , sleeping, eating, working, playing, colors =['orange' , 'g' , 'r' , 'b'])

plt.xlabel('Days')
plt.ylabel('Number of Hours spend')
plt.title("Day wise activity monitor")
plt.show()
```

```python
ax = plt.axes(projection = '3d')
zline = np.linspace(0,15,1000)
xline = np.sin(zline)
yline = np.cos(zline)

ax.plot3D(xline,yline,zline, 'purple');
```

```python
ax = plt.axes(projection = '3d')

zline = np.linspace(0,15,100)
xline = np.sin(zline)
yline = np.cos(zline)

ax.scatter3D(xline,yline,zline, c= 'green');
```

```python
ax = plt.axes(projection = '3d')

zline = np.linspace(0,15,100)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline,yline,zline , 'gray')

zdata = np.linspace(0,15,100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)

ax.scatter3D(xdata,ydata,zdata, c ='g' , cmap= "Greens");
```
