```python
from tkinter import *
from tkinter import messagebox

def calculate():
    try:
        num1 = float(num1_entry.get())
        num2 = float(num2_entry.get())
        operator = operator_entry.get()
        
        if operator == "+":
            result = num1 + num2
        elif operator == "-":
            result = num1 - num2
        elif operator == "*":
            result = num1 * num2
        elif operator == "/":
            if num2 != 0:
                result = num1 / num2
            else:
                messagebox.showerror("Error", "Division by zero is not allowed.")
                return
        else:
            messagebox.showerror("Error", "Invalid operator.")
            return
        
        result_label.configure(text=f"Result: {result}")
    except ValueError:
        messagebox.showerror("Error", "Invalid input. Please enter numbers only.")

r = Tk()
r.geometry("500x300")
r.configure(bg="khaki")
r.title("Simple Calculator")
r.resizable(0, 0)

num1_label = Label(r, text="Number 1:", font=("Segoe Script", 12))
num1_label.place(x=40, y=30)
num1_label.configure(bg="darkkhaki")
num1_entry = Entry(font="30")
num1_entry.configure(bg="palegoldenrod")
num1_entry.place(x=150, y=35)

num2_label = Label(r, text="Number 2:", font=("Segoe Script", 12))
num2_label.place(x=40, y=80)
num2_label.configure(bg="darkkhaki")
num2_entry = Entry(font="30")
num2_entry.configure(bg="palegoldenrod")
num2_entry.place(x=150, y=85)

operator_label = Label(r, text="Operator:", font=("Segoe Script", 12))
operator_label.place(x=40, y=130)
operator_label.configure(bg="darkkhaki")
operator_entry = Entry(font="30")
operator_entry.configure(bg="palegoldenrod")
operator_entry.place(x=150, y=135)

calculate_button = Button(r, text="Calculate", font="30", command=calculate)
calculate_button.place(x=40, y=200)
calculate_button.configure(bg="darkkhaki")

result_label = Label(r, text="Result:", font=("Segoe Script", 12))
result_label.place(x=40, y=250)
result_label.configure(bg="darkkhaki")

r.mainloop()
```
