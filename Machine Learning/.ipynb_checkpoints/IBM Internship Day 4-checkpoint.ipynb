```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy import array
```

```python
from pandas import read_csv
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df
```

```python
plt.scatter(df["Height"],df["Weight"])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(df[["Height"]],df["Weight"]) #fit is used to train the model
```

```python
slope = model.coef_
```

```python
intercept = model.intercept_
```

```python
print(slope)
print(intercept)
```

```python
model.predict([[70]])
```

```python
plt.plot(df["Height"], slope*df["Height"] +intercept, c = "black")
plt.scatter(df['Height'] , df["Weight"])
plt.xlabel("Height")
plt.ylabel("Weight")
```
